```python
df =pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df.head()
```

```python
df.dropna() #remove the rows or column having atleast one nan
```

```python
df1=pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, np.nan, np.nan],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df1
```

```python
df1.dropna(how ='all',axis =1)#removes  if all element of rows or column are  nan
```
